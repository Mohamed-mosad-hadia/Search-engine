```python
# Initialize an empty list to store cleaned job titles
clean_jobs = []

# Set of English stopwords
stop_words = set(stopwords.words('english'))

# Clean and tokenize job titles
for title in df['jobtitle']:
    # Convert to lowercase
    title_lower = title.lower()

    # Tokenize words
    tokens = word_tokenize(title_lower)

    # Remove stopwords and punctuation
    filtered_tokens = [word for word in tokens if word not in stop_words and word not in string.punctuation]

    # Join tokens back into a string
    clean_title = ' '.join(filtered_tokens)

    # Append to the list of cleaned job titles
    clean_jobs.append(clean_title)

# Add the cleaned job titles to the DataFrame
df['cleaned_jobtitle'] = clean_jobs

# Display the DataFrame with the cleaned job titles
print(df[['jobtitle', 'cleaned_jobtitle']].head())
```

# clustering using Kmean

```python
from sklearn.cluster import KMeans

# Assuming 'x' is your feature matrix
# Replace 'x' with your actual feature matrix

# Initialize an empty list to store the WCSS values
wcss = []

for i in range(1, 15):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, max_iter=600, n_init=1)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
#plotting the graph
plt.plot(range(1, 15), wcss, marker='o')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.show()
```

```python
true_k = 15
model = KMeans(n_clusters=true_k, init='k-means++', random_state=42, max_iter=600, n_init=1)
pred = model.fit_predict(X)
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names_out()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.metrics.pairwise import cosine_similarity

# Assuming vectorizer, svm, and df are pre-defined and loaded

def recommend_jobs(user_skills, vectorizer, svm, df):
    """
    Recommends jobs based on user skills.

    Parameters:
    user_skills (str): The skills input by the user.
    vectorizer (TfidfVectorizer): The vectorizer to transform text data.
    svm (SVC): The trained SVM model for job prediction.
    df (pd.DataFrame): The dataframe containing job listings and their associated labels.

    Returns:
    pd.DataFrame: Top 5 job listings with highest cosine similarity to the user skills.
    """

    # Transform the user input into a feature vector
    pred = vectorizer.transform([user_skills.lower()])

    # Predict the job category
    output = svm.predict(pred)
    print("You may look into " + output[0] + ' jobs')
    print("Here is a list of jobs under " + output[0])

    # Calculate cosine similarity and sort jobs
    cos = []
    labelData = df[df['label'] == output[0]].copy()

    for index, row in labelData.iterrows():
        skills = [row['skills']]
        skillVec = vectorizer.transform(skills)

        cos_lib = cosine_similarity(skillVec, pred)
        cos.append(cos_lib[0][0])

    labelData['cosine_similarity'] = cos
    top_5 = labelData.sort_values("cosine_similarity", ascending=False).head(5)[['advertiserurl', 'company', 'employmenttype_jobstatus', 'jobdescription', 'joblocation_address', 'jobtitle', 'skills', 'cleaned_jobtitle', 'label']]

    return top_5

# Example usage
# Assuming vectorizer, svm, and df are already defined and loaded
user_skills = input('Enter your skills: ')
top_jobs = recommend_jobs(user_skills, vectorizer, svm, df)
print(top_jobs.to_string(index=False))
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def get_user_input():
    return input('Enter your skills: ')

def predict_job_category(vectorizer, svm, userInput):
    pred = vectorizer.transform([userInput.lower()])
    return svm.predict(pred)[0]

def calculate_cosine_similarity(labelData, vectorizer, pred):
    skillVecs = vectorizer.transform(labelData['skills'].tolist())
    return cosine_similarity(skillVecs, pred).flatten()

def display_top_jobs(top_jobs):
    for index, row in top_jobs.iterrows():
        print(f"Company: {row['company']}")
        print(f"Job Title: {row['jobtitle']}")
        print(f"Location: {row['joblocation_address']}")
        print(f"Employment Type: {row['employmenttype_jobstatus']}")
        print(f"Job Description: {row['jobdescription']}\n")

def recommend_jobs(df, vectorizer, svm):
    userInput = get_user_input()
    job_category = predict_job_category(vectorizer, svm, userInput)
    print("You may look into " + job_category + ' jobs')
    print("Here is a list of jobs under " + job_category)

    labelData = df[df['label'] == job_category]
    pred = vectorizer.transform([userInput.lower()])
    labelData['cosine_similarity'] = calculate_cosine_similarity(labelData, vectorizer, pred)
    top_5 = labelData.sort_values("cosine_similarity", ascending=False).head(5)
    display_top_jobs(top_5)

# Call the function to get recommendations
recommend_jobs(df, vectorizer, svm)
```
